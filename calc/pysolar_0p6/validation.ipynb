```python
%matplotlib inline
```

```python
import pandas
pandas.set_option('display.mpl_style', 'default') # Make the graphs a bit prettier
df = pandas.read_csv('pysolar_v_usno.csv', index_col='timestamp')
```

### Azimuth error

```python
print('Mean error: {0} degrees'.format(df.az_error.mean()))
print('Standard deviation: {0} degrees'.format(df.az_error.std()))
print('Minimum error: {0} degrees'.format(df.az_error.min()))
print('Maximum error: {0} degrees'.format(df.az_error.max()))
```

### Altitude error

```python
print('Mean error: {0} degrees'.format(df.alt_error.mean()))
print('Standard deviation: {0} degrees'.format(df.alt_error.std()))
print('Minimum error: {0} degrees'.format(df.alt_error.min()))
print('Maximum error: {0} degrees'.format(df.alt_error.max()))
```

```python
alt_error_plot = df.plot(x='alt1', y='alt_error', marker='.', linestyle='')
alt_error_plot.set_xlabel('Altitude [degrees]')
alt_error_plot.set_ylabel('Error [degrees]')
alt_error_plot.set_title('Error in altitude between USNO and Pysolar')
```

```python
az_error_plot = df.plot(x='az1', y='az_error', marker='.', linestyle='')
az_error_plot.set_xlabel('Azimuth [degrees]')
az_error_plot.set_ylabel('Error [degrees]')
az_error_plot.set_title('Error in azimuth between USNO and Pysolar')
az_error_plot.set_xlim(0, 360)
```

```python
longitude_error_plot = df.plot('longitude', 'alt_error', 'longitude', 'az_error', marker='.', linestyle='')
longitude_error_plot.set_xlabel('Longitude [degrees]')
longitude_error_plot.set_ylabel('Error [degrees]')
longitude_error_plot.set_title('Error between USNO and Pysolar as a function of longitude')
longitude_error_plot.set_xlim(0, 360)
```

```python
latitude_error_plot = df.plot('latitude', 'alt_error', 'latitude', 'az_error', marker='.', linestyle='')
latitude_error_plot.set_xlabel('Latitude [degrees]')
latitude_error_plot.set_ylabel('Error [degrees]')
latitude_error_plot.set_title('Error between USNO and Pysolar as a function of latitude')
latitude_error_plot.set_xlim(-90, 90)
```

It may appear from the graph above that the latitude errors are worse in the northern hemisphere. I think that's not the case; we've just gathered around 10 times more data from the northern hemisphere because we didn't know how to retrieve data for the southern hemisphere from the USNO website until late 2013.
